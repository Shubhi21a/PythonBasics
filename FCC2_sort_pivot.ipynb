## Pivot

Pivot method reshapes data, does not do aggregations

```python
df = sales[['Product','Product_Category','Age_Group','Order_Quantity','Profit']]
df.head()
```

```python
#df1 = df.pivot(index='Product',columns='Product_Category',values='Profit')
# Error: Index contains duplicate entries, cannot reshape

# https://www.geeksforgeeks.org/python-pandas-dataframe-drop_duplicates/
df1 = df.drop_duplicates(subset='Product', keep='first')
print(df.shape)
df1.shape
```

```python
df1.head()
```

```python
df2 = df1.pivot(index='Product',columns='Product_Category',values='Profit')

df2.head()
```

## Pivot_Table
Supports Aggregation

```python
df.pivot_table(index='Age_Group',aggfunc="sum")
# Sums for all numeric columns, groups by index mentioned
```

```python
df.pivot_table(index='Age_Group',columns='Product_Category',aggfunc="sum")
# Sums for all numeric columns by mentioned column - Product_Category
```

```python
sales.pivot_table(index='Country',aggfunc="sum")
# Sums for all numeric columns
```

```python
sales.pivot_table(index='Country',columns='Product_Category',aggfunc="sum")
```

```python
sales.pivot_table(index='Country',values='Profit',aggfunc="sum")
# group profit, by country
```

```python
sales.pivot_table(index='Country',columns='Product_Category',values='Profit',aggfunc="sum")
# sums for mentioned column, by product_category
```

```python
sales.pivot_table(index='Country',columns='Product_Category',values=['Profit','Order_Quantity'],aggfunc="sum")
# sums for mentioned columns
```
